```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
import pandas as pd
import seaborn as sns

np.random.seed(0)

def monte_carlo_integrate(func, a, b, n_samples):
    x_samples = np.random.uniform(a, b, n_samples)
    return (b - a) * np.mean(func(x_samples))

# Two contrasting functions
def smooth(x):
    return np.sin(x)  # Smooth oscillation

def oscillatory(x):
    return 25*np.sin(25*x)  # Rapid oscillations

# True values
a, b = 0, np.pi
true_smooth = 2 #integrate.quad(smooth, a, b)
true_osc = 2 #integrate.quad(oscillatory, a, b)

# Visualization
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 8))

x = np.linspace(a, b, 10000)

# Functions
sns.lineplot(x=x, y=smooth(x), ax=ax1, color='b', linewidth=2)
ax1.fill_between(x, 0, smooth(x), alpha=0.3, color='b')
ax1.set_title(r'Smooth: $\sin(x)$')
ax1.grid(True, alpha=0.3)


sns.lineplot(x=x, y=oscillatory(x), ax=ax2, color='r', linewidth=2)
ax2.fill_between(x, 0, oscillatory(x), alpha=0.3, color='r')
ax2.set_title(r'Oscillatory: $25\sin(25x)$')
ax2.grid(True, alpha=0.3)

# Plot the estimates in ax3
ax3 = fig.add_subplot(221)
ax3.set_title('Estimate vs Sample Size')
sample_sizes = np.logspace(2, 5, 20).astype(int)
estimates_smooth = [monte_carlo_integrate(smooth, a, b, n) for n in sample_sizes]
estimates_osc = [monte_carlo_integrate(oscillatory, a, b, n) for n in sample_sizes]
ax3.plot(sample_sizes, estimates_smooth, 'b-o', label='Smooth')
ax3.plot(sample_sizes, estimates_osc, 'r-s', label='Oscillatory')
ax3.axhline(true_smooth, color='blue', linestyle='--', label='True Smooth Value')
ax3.axhline(true_osc, color='red', linestyle='--', label='True Oscillatory Value')
ax3.set_xscale('log')
ax3.set_xlabel('Sample Size')
ax3.set_ylabel('Integral Estimate')
ax3.legend()
ax3.grid(True, alpha=0.3)
# Convergence
samples = np.logspace(2, 5, 20).astype(int)
errors_smooth = [abs(monte_carlo_integrate(smooth, a, b, n) - true_smooth) for n in samples]
errors_osc = [abs(monte_carlo_integrate(oscillatory, a, b, n) - true_osc) for n in samples]

ax4.loglog(samples, errors_smooth, 'b-o', label='Smooth', linewidth=2)
ax4.loglog(samples, errors_osc, 'r-s', label='Oscillatory', linewidth=2)
ax4.set_xlabel('Sample Size')
ax4.set_ylabel('Absolute Error')
ax4.set_title('Error vs Sample Size')
ax4.legend()
ax4.grid(True, alpha=0.3)




# Multiple estimates at fixed sample size
# n_test = 10000
# estimates_smooth = [monte_carlo_integrate(smooth, a, b, n_test) for _ in range(25)]
# estimates_osc = [monte_carlo_integrate(oscillatory, a, b, n_test) for _ in range(25)]

# # Numerical comparison table as image in ax4
# # Prepare table data
# table_data = [
#     ["Function", "True Value", "Mean Est", "Std Dev", "Error"],
#     ["Smooth", f"{true_smooth:.6f}", f"{np.mean(estimates_smooth):.6f}", f"{np.std(estimates_smooth):.6f}", f"{abs(np.mean(estimates_smooth) - true_smooth):.6f}"],
#     ["Oscillatory", f"{true_osc:.6f}", f"{np.mean(estimates_osc):.6f}", f"{np.std(estimates_osc):.6f}", f"{abs(np.mean(estimates_osc) - true_osc):.6f}"],
# ]
# variance_ratio = np.std(estimates_osc)/np.std(estimates_smooth)

# # Create table
# ax4.axis('off')
# table = ax4.table(cellText=table_data,
#                   cellLoc='center',
#                   loc='center',
#                   colWidths=[0.18]*5)
# table.auto_set_font_size(False)
# table.set_fontsize(12)
# table.scale(1.2, 1.2)

# ax4.set_title(f"Variance ratio (Osc/Smooth): {variance_ratio:.1f}x higher", fontsize=13)

plt.tight_layout()
plt.show()
```

```python
#| label: fig-multidimensional
#| fig-cap: Monte Carlo advantage in higher dimensions

def multidimensional_example():
    """Demonstrate MC integration in higher dimensions"""
    
    # Function: exp(-||x||) over unit hypercube
    def f_nd(points):
        return np.exp(-np.sum(points**2, axis=1))
    
    dimensions = [1, 2, 3, 4, 5]
    n_samples = 2500
    
    results = []
    
    for dim in dimensions:
        estimates = []
        for _ in range(100):  # Multiple runs
            # Sample points in [0,1]^d
            points = np.random.uniform(0, 1, (n_samples, dim))
            f_values = f_nd(points)
            estimate = np.mean(f_values)  # Volume of unit cube is 1
            estimates.append(estimate)
        
        mean_est = np.mean(estimates)
        std_est = np.std(estimates)
        
        results.append({
            'Dimension': dim,
            'Estimate': mean_est,
            'Std Error': std_est,
            'Rel Error': std_est/mean_est if mean_est > 0 else np.inf
        })
    
    df = pd.DataFrame(results)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
    
    # Estimates by dimension with error bars
    ax1.errorbar(df['Dimension'], df['Estimate'], 
                yerr=df['Std Error'], 
                fmt='bo-', linewidth=2, markersize=8, capsize=5, capthick=2)
    ax1.set_xlabel('Dimension')
    ax1.set_ylabel('Integral Estimate')
    ax1.set_title('Integration by Dimension')
    ax1.grid(True, alpha=0.3)
    
    # Relative error with error bars showing uncertainty in the error estimate
    ax2.semilogy(df['Dimension'], df['Rel Error'], 'ro-', linewidth=2, markersize=8)
    ax2.set_xlabel('Dimension')
    ax2.set_ylabel('Relative Error')
    ax2.set_title('Relative Error vs Dimension')
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("\nMultidimensional Integration Results From 2500 samples:")
    print(df.to_string(index=False, float_format='%.6f'))

multidimensional_example()
```
